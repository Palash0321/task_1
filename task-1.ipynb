### Handling missing values

```python
df['director'].fillna('Unknown',inplace=True)
```

```python
df['cast'].fillna('Unknown',inplace=True)
```

```python
df['country'].fillna('Unknown',inplace=True)
```

```python
df['date_added'].fillna('Unknown',inplace=True)
```

```python
df['rating'].fillna('Unknown',inplace=True)
df['duration'].fillna('Unknown',inplace=True)
```

```python
df
```

```python
print("----No missing Values Found----")
```

### Remove Duplicate Rows:

```python
print("\n--- Removing Duplicate Rows ---")
# This will remove duplicates based on 'show_id' if any exist.
df.drop_duplicates(subset=['show_id'], inplace=True)
print("Duplicate rows based on 'show_id' removed (if any were present).")
print("Number of rows after duplicate removal:", len(df))
```

```python
df
```

### Standardize Text Values:

```python
print("\n--- Standardizing Text Values ---")
# Standardize 'type' to lowercase
df['type'] = df['type'].str.lower()
print(f"Unique 'type' values after standardization: {df['type'].unique()}")
```

```python
df
```

### Standardize 'listed_in' (genres) - strip whitespace, split, strip again, then join


```python
df['listed_in'] = df['listed_in'].apply(lambda x: ', '.join([s.strip().title() for s in x.split(',')]))
print(f"Example 'listed_in' values after standardization (first 5 unique): {df['listed_in'].unique()[:5]}")
```

```python
df
```

### Standardize 'director', 'cast', 'country', 'rating' to Title Case and strip whitespace


```python
text_cols_to_standardize = ['director', 'cast', 'country', 'rating']
for col in text_cols_to_standardize:
 df[col] = df[col].str.strip().str.title()
 print(f"Column '{col}' standardized to Title Case and stripped whitespace.")
```

```python
df
```

```python
print("\n--- Final DataFrame Info After All Cleaning Steps ---")
print(df.info())

# Ensure 'release_year' is integer type
df['release_year'] = df['release_year'].astype(int)
print("\n'release_year' column ensured to be integer type.")
```
